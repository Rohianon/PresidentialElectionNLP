```python
import re
import csv
import pandas as pd
```

```python
def parse_file(filepath):
    """
    This function parses through text from given paths.
    
    Input:
    filepath : string
        File path for file_object to be parsed
    
    Returns:
    data: a Pandas Dataframe
        Parsed Data.
    """
    #empty list for data collection
    data = []
    #open file and read it line after line
    with open(filepath, 'r', encoding='utf8') as some_file:
        reader = csv.reader(some_file)
        
        for row in reader:
            data.append(row)
        some_file.close()
            
        #using pandas to create a dataframe
        headers = data[0]
        # print(headers)
        data = pd.DataFrame(data[1:], columns = headers)
        # export to csv 
        export_csv = data.to_csv(r'Us_2020_Presidential_Debate.csv', index=None, header=True)
    return data
```
